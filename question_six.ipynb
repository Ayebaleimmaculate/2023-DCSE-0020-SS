```python

import numpy as np

#average number of arrivals per day
lambda_ = 5

#1000 random samples from the Poisson distribution
poisson_samples = np.random.poisson(lambda_, 1000)

#for the first 10 samples
print("First 10 samples:", poisson_samples[:10])

#total number of samples
print("Total number of samples:", len(poisson_samples))

```

```python

#The mean and variance of the generated samples
mean = np.mean(poisson_samples)
variance = np.var(poisson_samples)

# Print the results
print("Mean of the samples:", mean)
print("Variance of the samples:", variance)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot the histogram
plt.hist(poisson_samples, bins=20, edgecolor='black', alpha=0.7)

# Add title and labels
plt.title(f'Histogram of Poisson Distribution (λ = {lambda_})')
plt.xlabel('Number of arrivals')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
